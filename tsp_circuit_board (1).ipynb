```python
import pandas as pd
```

```python
path = '/content/LOG_20210101_001253_025162(藤井寺IC)_L5.xlsx'
data = pd.ExcelFile(path)
sheets = data.sheet_names
df = pd.DataFrame()
for s in sheets[:]:
  tmp = data.parse(s, skiprows=[0, 1, 2, 3, 4, 5], index_col=0)
  df = pd.concat([df, tmp])
```

```python
df.reset_index(drop=True, inplace=True)
```

```python
# 7040910
idx_v = (df[df['ETC\n一連番号'] == 7040910].index[0], df[df['ETC\n一連番号'] == 7040910].index[-1])
df_v = df.iloc[idx_v[0]: idx_v[1]+1, :]
df_v.reset_index(drop=True, inplace=True)
df_v
```

```python
idx = (df_v[df_v['全イベント'] == '車検変化'].index[0], df_v[df_v['全イベント'] == '車検変化'].index[-1])
# print(idx)
df_v2 = df_v.iloc[idx[0]: idx[1]+1, :]
df_v2.reset_index(drop=True, inplace=True)
df_v2
```

```python
cols = ['全イベント', 'Unnamed: 38',
       'Unnamed: 39', 'Unnamed: 40', 'Unnamed: 41', 'Unnamed: 42',
       'Unnamed: 43', 'Unnamed: 44', 'Unnamed: 45', 'Unnamed: 46',
       'Unnamed: 47']
df_v2[cols]
```

```python
sub_df_v2 = df_v2[df_v2['全イベント']=='車検変化']
sub_df_v2.reset_index(drop=True, inplace=True)
sub_v2 = {'nodes': [], 'combos': [], 'edges': []}
sub_v2['combos'].append({'id': 'n.車検変化', 'label': '車検変化'})
for i in sub_df_v2['Unnamed: 39'].value_counts().index.tolist():
  sub_v2['nodes'].append({'id': i, 'label': i, 'comboId': 'n.車検変化'})
for i in range(len(sub_df_v2)-1):
  # print(i)
  # print(sub_df_v2.loc[i, 'Unnamed: 39'], sub_df_v2.loc[i+1, 'Unnamed: 39'])
  sub_v2['edges'].append({'source': sub_df_v2.loc[i, 'Unnamed: 39'], 'target': sub_df_v2.loc[i+1, 'Unnamed: 39']})
```

```python
import json
with open('data-sub-v3.1.json', 'wb') as fp:
    fp.write(json.dumps(sub_v2, ensure_ascii=False).encode("utf8"))
```

```python
df_v2['combo'] = ''
key_words = ['ＮＰ読取', '軸検知', '不正防止', '料金所サ', 'ＥＴＣ処理', '第１アンテナ', 'ＤＳＲＣ通信', \
                '第２アンテナ', '車検変化', 'Ｓ１割込み', '路側表示', '車監装', '発進制御', '簡易操作', \
                    '車載器指示', '車両台数', '重量測定装置']
reject_combo = []
for k in key_words:
  row_indexer = df_v2[df_v2['全イベント'].str.contains(k)].index
  if len(row_indexer) == 0:
    reject_combo.append(k)
  df_v2.loc[row_indexer, 'combo'] = 'c.'+k
```

```python
v2 = {'nodes': [], 'combos': [], 'edges': []}
for i, label in enumerate([i for i in key_words if i not in reject_combo]):
  v2['combos'].append({'id': 'c.'+ label, 'label': label})
```
